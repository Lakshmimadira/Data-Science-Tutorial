```python
import matplotlib.pyplot as plt
#creating a plot
plt.plot([1,2,3,4],[10,20,25,30])

#customizing axis limits and lables
plt.xlim(0,5)
plt.ylim(0,35)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('sample plot')
plt.show()
```

```python
# to generate a sine wave
from matplotlib import pyplot as plt
import numpy as np
import math  # needed for definiton of pi
x=np.arange(0,math.pi*2,0.05)
y=(np.sin(x))*2
#print(x)
#print(y)
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title("sine wave")
plt.show()
```

```python
import matplotlib.pyplot as plt
#example data
year=[1,2,3]
veg_rate=[2,4,6]
milk_rate=[1,3,5]
fruit_rate=[3,6,9]

#plotting the data
line1,=plt.plot(year,veg_rate)
line2,=plt.plot(year,milk_rate)
line3,=plt.plot(year,fruit_rate)

plt.legend([line1,line2,line3],['vegetables','milk','fruit']);
plt.xlabel("years");
plt.ylabel("rate");
plt.show()
print("succesfully placed a legend on the axes")
```

```python
# bar chart
import matplotlib.pyplot as plt
categories =['lion','horse','kangaroo']
values=[15,34,20]
plt.bar(categories,values,color='skyblue')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('basic vertical bar graph')
plt.show()
```

```python
import matplotlib.pyplot as plt
categories =['ferrari','benz','honda']
values=[48,28,35]
plt.barh(categories,values,color=['orange','white','green'])
plt.xlabel('categories')
plt.ylabel('values')
plt.title('basic vertical bar graph')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
categories=['lion','horse','kangaroo']
values1=[15,24,30]
values2=[20,18,25]
bar_width=0.35
bar_position1=np.arange(len(categories))
bar_position2=bar_position1+bar_width
plt.bar(bar_position1,values1,width=bar_width,label='2021',color='skyblue')
plt.bar(bar_position2,values2,width=bar_width,label='2022',color='yellow')
plt.xlabel('categories')
plt.ylabel('values')
plt.title("grouped bar chart")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
categories=['Transport','Income Tax','Housing']
values1=[15,24,30]
values2=[20,18,25]
plt.bar(categories,values1,label='First Half',color='skyblue')
plt.bar(categories,values2,bottom=values1,label='Second Half')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Stacked bar graph')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x=[70,50,100,70,50,100,50,50,50,70,50]
plt.hist(x,orientation="vertical",color='green')
plt.title("Mark Analysis")
plt.ylabel("No.of students")
plt.xlabel("Marks")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
exam_scores =np.random.randint(0,100,150)
plt.hist(exam_scores,bins=20,cumulative=True,color='skyblue',edgecolor='k')
plt.xlabel('Exam Scores')
plt.ylabel('cumulative no.of students')
plt.title('Cumulative Histogram of exam scores')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
marks1=np.array([2,4,5,7,9,10,11,13,14,15])
marks2=np.array([6,7,8,10,11,12,13,14,15,16])
plt.hist([marks1,marks2],bins=10,stacked=True,color=['skyblue','salmon'],edgecolor='k')
plt.xlabel('Marks')
plt.ylabel('Count')
plt.title('Stacked Histogram with multiple dataset')
plt.legend(['Mark Dataset1','Mark Dataset2'])
plt.show()
```
