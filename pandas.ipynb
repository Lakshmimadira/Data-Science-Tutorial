```python
import pandas as pd
# Create a DataFrame
data = {
    'Age': [23, 45, 12, 35, 46],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Salary': [70000, 80000, 15000, 50000, 90000]
}
df = pd.DataFrame(data)
print("DataFrame:")
print(df)
# Data manipulation
# 1. Selecting a column
print("\nAges:")
print(df['Age'])
# 2. Filtering rows
print("\nFiltered DataFrame (Age > 30):")
filtered_df = df[df['Age'] > 30]
print(filtered_df)
# 3. Adding a new column
df['Tax'] = df['Salary'] * 0.1
print("\nDataFrame with Tax column:")
print(df)
# 4. Grouping and aggregation
print("\nMean Salary by Age Group:")
print(df.groupby('Age')['Salary'].mean())
# 5. Handling missing values
df_with_nan = df.copy()
df_with_nan.loc[2, 'Salary'] = None  # Introduce a NaN value
print("\nDataFrame with NaN:")
print(df_with_nan)
print("\nDataFrame after filling NaN with median salary:")
df_with_nan['Salary'] = df_with_nan['Salary'].fillna(df_with_nan['Salary'].median())
print(df_with_nan)

```

```python
#series

import pandas as pd
a=[1,7,2]
myvar=pd.Series(a)
print(a)
print(myvar)
print('--------')


myvar=pd.Series(a)
print(a)
print(myvar)
print('--------')
myvar2=pd.Series(a,index=["x","y","z"])
print(myvar2)
print(myvar2["y"])
print('-----------')

#dictionary
calories={"day1":420,"day2":300,"day3":390}
myvar=pd.Series(calories)
print(myvar)
print('--------')


#series generated from scalar value
series=pd.Series(5,index=[0,1,2,3])
print(series)
print('--------')

#integer series
int_series=pd.Series([1,2,3,4])
print(int_series)
print('-------')

#float series
float_series=pd.Series([1.1,2.2,3.3,4.4])
print(float_series)
print('-------')

#string series
string_series=pd.Series(['a','b','c','d'])
print(string_series)
print('-------')


#label-based indexing
data={'a':10,'b':20,'c':30}
series=pd.Series(data)
print(series['a'])
print(series['a'],['c'])
print('--------')

#boolean indexing
data=[10,20,30,40,50]
series=pd.Series(data)
#access elements greater than 25
print(series[series>25])
print('-----------')


#dataframe
data={
   'Name':['Alice','Bob','Charlie'],
   'Age':[25,30,35],
   'City':['New York','Los Angels','Chicago']
}
df=pd.DataFrame(data)
print(df)
print('--------')
#selecting columns
print(df.Name)
print(df['Name'])
print(df[['Name','City']])
print('--------')
```

```python
import pandas as pd
data={
   'Name':['Alice','Bob','Charlie','Alice','Bob'],
   'Age':[25,30,35,25,30],
   'City':['New York','San franscico','Chicago','Los Angels','San franscico'],
   'Salary':[70000,80000,120000,700000,800000]
}
df=pd.DataFrame(data)
print(df)
print('------')
#groupby
grouped=df.groupby('Name')
print(grouped.groups)
#sum of salaries for each group
salary_sum=grouped['Salary'].sum()
print(salary_sum)
age_mean=grouped['Age'].mean()
print(age_mean)
```

```python
#dictionary containing employee data
import pandas as pd
data1={'Name':['jai','prince','gaurav','anju'],
       'Age':[27,24,22,32],
       'Address':['nagapur','kanpur','allahabad','kannuaj'],
       'Qualification':['msc','na','mca','phd']}
data2={'Name':['abhi','ayushi','dhiraj','hitesh'],
       'Age':[17,14,12,52],
       'Address':['nagapur','kanpur','allahabad','kannuaj'],
       'Qualification':['btech','b.a','bcom','b.hons'],
       'Salary':[1000,2000,3000,4000]}
#dictionary into dataframe
df=pd.DataFrame(data1,index=[0,1,2,3])
df1=pd.DataFrame(data2,index=[2,3,6,7])
print(df,"\n\n", df1)
res1=pd.concat([df,df1])
res1
res2=pd.concat([df,df1],axis=1,join='inner')
res2
```

```python
import pandas as pd
data1={'Key':['k0','k1','k2','k3'],
       'Name':['Jai','Prince','Gaurav','Anuj'],
       'Age':[27,24,22,32]}
data2={'Key':['k0','k1','k2','k3'],
       'Address':['Nagpur','Kanpur','allahabad','kannuaj'],
       'Qualification':['btech','b.a','bcom','b.horns']}
df=pd.DataFrame(data1)
df1=pd.DataFrame(data2)
print(df,"\n\n",df1)
res=pd.merge(df,df1,on='Key')
res

data1={'Key':['k0','k1','k2','k3'],
       'Key1':['k0','k1','k2','k3'],
       'Name':['Jai','Prince','Gaurav','Anuj'],
       'Age':[27,24,22,32]}
data2={'Key':['k0','k1','k2','k3'],
       'Key1':['k0','k0','k0','k0'],
       'Address':['Nagpur','Kanpur','allahabad','kannuaj'],
       'Qualification':['btech','b.a','bcom','b.horns']}
df=pd.DataFrame(data1)
df1=pd.DataFrame(data2)
print(df,"\n\n",df1)
res1=pd.merge(df,df1,on=['Key','Key1'])
res1
```
